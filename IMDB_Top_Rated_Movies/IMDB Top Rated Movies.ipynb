```python
from selenium import webdriver
from bs4 import BeautifulSoup
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import ElementNotVisibleException
import time

url = "https://www.imdb.com/chart/top/?sort=ir,desc&mode=simple&page=1"
driver = webdriver.Chrome(executable_path=r"C:\Users\shrut\Downloads\chromedriver.exe")
driver.get(url)

rank=[]; title=[]; year=[]

ratings=[]; user_ratings=[]

running_time=[]; genre=[]; storyline=[];
directors=[]; writers=[]; stars=[]


try:

    soup = BeautifulSoup(driver.page_source,'html.parser')


    cnt=0
    for i in soup.findAll("td",class_="titleColumn"):


    #   Rank
        cnt+=1
        print(cnt)
        rank.append(cnt)

        '''----------------------------------------------------------------------'''
    #   Title
        t=i.find("a").text.strip()
        title.append(t)
        print(t)


        '''----------------------------------------------------------------------'''
    #   Year
        y=i.find("span").text[1:-1].strip()
        year.append(y)
        print(y)


        '''********************************************************************************************'''      
#           Getting down to individal movies
        movie = driver.find_element_by_partial_link_text(t)
        time.sleep(2)
        movie.click()
        time.sleep(2)


        soup1 = BeautifulSoup(driver.page_source,'html.parser')

        '''----------------------------------------------------------------------'''
#           Rating out of 10
        for j in soup1.find("span", itemprop="ratingValue"):
            print(j)
            ratings.append(j)

        '''----------------------------------------------------------------------'''
#           User Rating Count
        for j in soup1.find("span", itemprop="ratingCount"):

            print(j)
            user_ratings.append(j)

        '''----------------------------------------------------------------------'''
#           Runtime
        r_time=soup1.find("time")
        running_time.append(r_time.text.strip())
        print(running_time[cnt-1])

        '''----------------------------------------------------------------------'''
#           Genre
        g=[]

        for j in soup1.find("div",class_="subtext").findAll("a")[:-1]:
            g.append(j.text.strip())

        genre.append(g)
        print(genre[cnt-1])

        '''----------------------------------------------------------------------'''
#           Summary
        div=soup1.find("div",class_="inline canwrap")
        summary=div.find("span").text.strip()
        print(summary,"\n")
        storyline.append(summary)

        '''----------------------------------------------------------------------'''
#           MainCast

        print("Stars-")
        for star in soup1.findAll("div",class_="credit_summary_item")[2].findAll("a")[:-1]:
            print(star.text)
            stars.append(star.text.strip())

        print("\n")

        '''********************************************************************************************''' 
#           Diving Deeper for Cast and Crew Details

        cast_and_crew = driver.find_element_by_partial_link_text("See full cast & crew")
        time.sleep(2)
        cast_and_crew.click()
        time.sleep(2)

        soup2=BeautifulSoup(driver.page_source,'html.parser')

        '''----------------------------------------------------------------------'''
#           Director's Name
        d=""
        tagn=soup2.find("div",id="fullcredits_content")  
        for dire in tagn.findAll("table",class_="simpleTable simpleCreditsTable")[0].findAll("td",class_="name"):
            d=d+(dire.text.strip())+"\n"

        directors.append(d)
        print("Directors:\n",d)

        '''----------------------------------------------------------------------'''
#           Writer's Name
        w=""
        wtagn=soup2.find("div",id="fullcredits_content") 
        for writer in wtagn.findAll("table",class_="simpleTable simpleCreditsTable")[1].findAll("td",class_="name"):
            w=w+(writer.text.strip())+"\n"

        writers.append(w)
        print("Writers:\n",w)

        '''********************************************************************************************''' 

        print("\n")

        driver.back()
        driver.back()


except NoSuchElementException:


    print("Reached bottom of page")
    driver.close()
    break

except ElementNotVisibleException:
    print("Reached bottom of page")
    driver.close()
    break
        
driver.close()
```

```python
rank=rank[:250]
title=title[:250]
year=year[:250]
ratings=ratings[:250]
user_ratings=user_ratings[:250]
running_time=running_time[:250]
genre=genre[:250]
storyline=storyline[:250]
directors=directors[:250]
stars=stars[:250]
```

```python
import pandas as pd
df=pd.DataFrame({
    
    "Rank":rank,
    "Title":title,
    "Year":year,
    "Ratings":ratings,
    "User Ratings":user_ratings,
    "Running Time":running_time,
    "Genre":genre,
    "StoryLine":storyline,
    "Directors":directors,
    "Stars":stars
    
})
```

```python
df.to_csv("IMDB_Top_Rated_Movies.csv")
```
