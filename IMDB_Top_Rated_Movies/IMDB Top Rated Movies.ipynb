```python
from selenium import webdriver
from bs4 import BeautifulSoup
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import ElementNotVisibleException
import pandas as pd
import time

url = "https://www.imdb.com/chart/top/?sort=ir,desc&mode=simple&page=1"
driver = webdriver.Chrome(executable_path=r"C:\Users\shrut\Downloads\chromedriver.exe")
driver.get(url)

def get_title(soup):
 
        t=soup.find("a").text.strip()
        print(t)
        return t
    
def get_year(soup):
    
        y=soup.find("span").text[1:-1].strip()
        print(y)
        return y

def get_ratings(soup):
        
        rat=soup.find("span", itemprop="ratingValue").text.strip()
        ratCnt=soup.find("span", itemprop="ratingCount").text.strip()
        
        print(rat,"\n",ratCnt)
        return rat,ratCnt
                    
def get_runtime(soup):
    
        r_time=soup.find("time")
        print(r_time.text.strip())
        return r_time.text.strip()
       
def get_genre(soup):
    
        g=[]

        for j in soup.find("div",class_="subtext").findAll("a")[:-1]:
            g.append(j.text.strip())
        
        print(g)

        return g
    
def get_summary(soup):
    
    div=soup.find("div",class_="inline canwrap")
    summary=div.find("span").text.strip()
    print(summary)
    return summary
    
    
def get_main_cast(soup):
    
    stars=""
    print("Stars-")
    for star in soup.findAll("div",class_="credit_summary_item")[2].findAll("a")[:-1]:
        stars=stars+(star.text.strip())+"\n"
    print(stars)
    return stars
    
def get_directors(soup):
    
    directors=""
    tagn=soup.find("div",id="fullcredits_content")  
    for dire in tagn.findAll("table",class_="simpleTable simpleCreditsTable")[0].findAll("td",class_="name"):
        directors=directors+(dire.text.strip())+"\n"
    
    print("Directors:",directors)
    return directors
 
def get_writers(soup):
    
    writers=""
    wtagn=soup.find("div",id="fullcredits_content") 
    for writer in wtagn.findAll("table",class_="simpleTable simpleCreditsTable")[1].findAll("td",class_="name"):
        writers=writers+(writer.text.strip())+"\n"

    print("Writers:\n",writers)
    return writers
    

    
                
def get_movies():
    
        rank=[]; title=[]; year=[]

        ratings=[]; user_ratings=[]

        running_time=[]; genre=[]; storyline=[]
        
        directors=[]; writers=[]; stars=[]
        
        
        try:

            soup = BeautifulSoup(driver.page_source,'html.parser')


            cnt=0
            for i in soup.findAll("td",class_="titleColumn"):


#               Rank
                cnt+=1
                print(cnt)
                rank.append(cnt)

                '''----------------------------------------------------------------------'''
                
#               Title
                title.append(get_title(i))


                '''----------------------------------------------------------------------'''
#               Year
                year.append(get_year(i))



                '''********************************************************************************************''' 
        
#               Getting down to individal movies
                movie = driver.find_element_by_partial_link_text(get_title(i))
                time.sleep(2)
                movie.click()
                time.sleep(2)


                soup1 = BeautifulSoup(driver.page_source,'html.parser')
        
                '''----------------------------------------------------------------------'''
            
#               Rating out of 10
                ratings.append(get_ratings(soup1)[0])

                '''----------------------------------------------------------------------'''
        
#               User Rating Count
                user_ratings.append(get_ratings(soup1)[1])

                '''----------------------------------------------------------------------'''
        
#               Runtime
                running_time.append(get_runtime(soup1))
              

                '''----------------------------------------------------------------------'''
            
#               Genre
                genre.append(get_genre(soup1))
               
                '''----------------------------------------------------------------------'''
            
#               Summary
               
                storyline.append(get_summary(soup1))

                '''----------------------------------------------------------------------'''
            
#               MainCast

                stars.append(get_main_cast(soup1))
    

                '''********************************************************************************************''' 
        
#               Diving Deeper for Cast and Crew Details

                cast_and_crew = driver.find_element_by_partial_link_text("See full cast & crew")
                time.sleep(2)
                cast_and_crew.click()
                time.sleep(2)

                soup2=BeautifulSoup(driver.page_source,'html.parser')

                '''----------------------------------------------------------------------'''
        #           Director's Name
               
                directors.append(get_directors(soup2))
         

                '''----------------------------------------------------------------------'''
        #           Writer's Name
               
                writers.append(get_writers(soup2))

                '''********************************************************************************************''' 

                print("\n")

                driver.back()
                driver.back()


        except NoSuchElementException:


            print("Reached bottom of page")
            driver.close()

        except ElementNotVisibleException:
            
            print("Reached bottom of page")
            driver.close()   
            
     

        
        driver.close()
        
        df=pd.DataFrame({

            "Rank":rank,
            "Title":title,
            "Year":year,
            "Ratings":ratings,
            "User Ratings":user_ratings,
            "Running Time":running_time,
            "Genre":genre,
            "StoryLine":storyline,
            "Directors":directors,
            "Stars":stars

        })
        
        return df
        
        
df=get_movies()
```
